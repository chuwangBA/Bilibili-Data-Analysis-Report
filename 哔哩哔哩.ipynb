```python
#对1200个视频数据集的多元回归建模
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
df.info()
```

```python
df=df[df['分享数']<400000]
df=df[df['综合得分']<2500000]
df=df[df['评论数']<100000]
df=df[df['点赞数']<1200000]
df=df[df['投币数']<1400000]
Xvar=df[['弹幕数','点赞数','综合得分','分享数','收藏数','评论数','投币数']]
Yvar=df['播放量_y']

for i in Xvar.columns:
    plt.xlabel(i)
    plt.ylabel('播放量')
    plt.scatter(Xvar[i],Yvar)
    plt.show()
```

```python
t=df[['弹幕数','点赞数','综合得分','分享数','收藏数','评论数','投币数','播放量_y']]
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(np.abs(t.corr()),annot=True)
plt.show()
```

```python
Xstd=std.fit_transform(Xvar)
Y=Yvar.values
X=Xstd
X=sm.add_constant(X)
lm=sm.OLS(Y,X).fit()
print(lm.summary())
#弹幕数和综合得分没有通过检验
```

```python
#PCA主成分分析
from sklearn.decomposition import PCA
pca_model=PCA(n_components=2)
X=Xstd
Y=Yvar.values
pca_model.fit(X)
X_pca=pca_model.transform(X)
X_pca=sm.add_constant(X_pca)
lm=sm.OLS(Y,X_pca).fit()
print(lm.summary())
```

```python
np.round(pca_model.components_,2)
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=0.5)
model.fit(Xstd,Yvar)
print('回归系数：',np.round(model.coef_,0))
predicted_X=model.predict(Xstd) #预测公式
plt.scatter(range(len(Xvar)),Yvar-predicted_X,c='r')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.show()
```
